## **Q8**

```python
sankranti_campaign.loc[:,'rev_before_promo'] = sankranti_campaign['base_price(before_promo)'] * sankranti_campaign['quantity_sold(before_promo)']
sankranti_campaign.loc[:,'rev_after_promo'] = sankranti_campaign['base_price(after_promo)'] * sankranti_campaign['quantity_sold(after_promo)']
sankranti_campaign
```

```python
sankranti_campaign.loc[:,'IR%'] = (sankranti_campaign['rev_after_promo'] - sankranti_campaign['rev_before_promo'])/sankranti_campaign['rev_before_promo'] * 100
sankranti_campaign.loc[:,'IR%'] = np.round(sankranti_campaign['IR%'], 2)
sankranti_campaign
```

```python
grouped = sankranti_campaign.groupby('product_code')
grouped
```

```python
for code, group in grouped:
    print(code)
    print(group)
    print('----------------------')
```

```python
# Aggregate revenue and quantity sold by product_code
aggregated_campaign = sankranti_campaign.groupby('product_code').agg(
    rev_before_promo=('rev_before_promo', 'sum'),
    rev_after_promo=('rev_after_promo', 'sum')
).reset_index()

# Calculate IR% for each product after aggregation
aggregated_campaign['IR%'] = (
    (aggregated_campaign['rev_after_promo'] - aggregated_campaign['rev_before_promo']) 
    / aggregated_campaign['rev_before_promo']
) * 100

# Round IR% to 2 decimal places
aggregated_campaign['IR%'] = np.round(aggregated_campaign['IR%'], 2)

# Find the product with the highest IR%
highest_ir_percentage = aggregated_campaign['IR%'].max()
product_code = aggregated_campaign[aggregated_campaign['IR%'] == highest_ir_percentage]['product_code'].iloc[0]

# Lookup the product name using the product_code
product_name = df_products[df_products['product_code'] == product_code]['product_name'].iloc[0]

# Print results
print('Highest Incremental Revenue Percentage for Sankranti Campaign:', highest_ir_percentage)
print(f'The Product which recorded highest Incremental Revenue Percentage (IR%) during the Sankranti campaign: {product_name} with IR%: {highest_ir_percentage}')
```

```python
print(f'The promo types which had both a negative Incremental Revenue Percentage (IR%) and Incremental Sold Units Percentage (ISU%) during the Sankranti campaign: {' and '.join(promo_types)}.')
```

```python
# Aggregate revenue and quantity sold by product_code
aggregated_campaign = sankranti_campaign.groupby('promo_type').agg(
    quantity_sold_before_promo=('quantity_sold(before_promo)', 'sum'),
    quantity_sold_after_promo=('quantity_sold(after_promo)', 'sum'),
    rev_before_promo=('rev_before_promo', 'sum'),
    rev_after_promo=('rev_after_promo', 'sum')
).reset_index()

# Calculate IR% for each product after aggregation
aggregated_campaign['ISU%'] = (
    (aggregated_campaign['quantity_sold_after_promo'] - aggregated_campaign['quantity_sold_before_promo']) 
    / aggregated_campaign['quantity_sold_before_promo']
) * 100

# Round IR% to 2 decimal places
aggregated_campaign['ISU%'] = np.round(aggregated_campaign['ISU%'], 2)

# Calculate IR% for each product after aggregation
aggregated_campaign['IR%'] = (
    (aggregated_campaign['rev_after_promo'] - aggregated_campaign['rev_before_promo']) 
    / aggregated_campaign['rev_before_promo']
) * 100

# Round IR% to 2 decimal places
aggregated_campaign['IR%'] = np.round(aggregated_campaign['IR%'], 2)
```
